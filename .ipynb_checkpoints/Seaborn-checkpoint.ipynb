```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Distribution plot

## Histogram 
#### best for univariate (one variable)

```python
#histogram best for univariate (one variable)
sns.distplot(tips['total_bill'])
#line is kde
```

```python
#distribution of total bill counts
#y counts
#x bins
sns.distplot(tips['total_bill'],kde=False,bins=40)
```

## Joint plot 
#### best for bivariate data (combine 2 distribution plot)

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#kind change on whats going on joint plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

## Pairplot
#### Plot pairweis relationship entire dataframe (at least for numerical), support color hue argument on categorical columns

```python
sns.pairplot(tips)
```

```python
#passing hue in categorical column
sns.pairplot(tips,hue='sex')
```

```python
#palette to specify a palette to color figure
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

## Rugplot
#### draw a dashmark for every single point along the distribution line

```python
sns.rugplot(tips['total_bill'])
```

## KDE kernel density estimation plot

```python
sns.kdeplot(tips['total_bill'])
```

# Categorical plot

```python
tips.head()
```

## Barplot
#### allows aggregate categorical data by function

```python
import numpy as np
```

```python
#x axis category, y axis numerical
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
#estimator is parameter for statistics aggregatio, by default estimator is mean
#e.g np.std = estimator standard deviation
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

## Countplot
#### estimator default counting the number of occurences, thats why only x value

```python
sns.countplot(x='sex',data=tips)
```

## Boxplot
#### shown distribution categorical data

```python
#not only days category on x axis, we can also add smoker category with hue
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

# Violin plot
#### same as box plot, show distributinn of categorical data

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

## Strip plot
#### draw scatter plot where one var is categorical

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex')
```
